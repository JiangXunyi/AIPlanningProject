```python
import pandas as pd
import numpy as np
```

```python
file_name = 'logs.csv'

df = pd.read_csv(file_name)
print(df.head())
```

```python
len(df)
```

```python
import matplotlib.pyplot as plt


# Assuming 'loss' is the name of the column in your DataFrame
loss0_values = df['loss_0']
loss0_values = loss0_values.apply(pd.to_numeric, errors='coerce')
loss0_values.dropna(inplace=True)
loss0_values = loss0_values.drop(loss0_values.index[:500])
# Plotting the curve
plt.plot(loss0_values)
plt.yticks([min(loss0_values), max(loss0_values)])
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss Curve')

plt.show()
```
